```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv("Oversampled_data.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['target'].unique()
```

```python
dataset['target']=dataset['target'].replace(to_replace = {'ckd': 'ckd' , 'ckd\t': 'ckd', 'notckd': 'notckd'})
dataset['target'].unique()
```

```python
dataset.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
col = ['rbc', 'pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane', 'target']
for i in col:
    dataset[i] = le.fit_transform(dataset[i]).astype(int)
```

```python
dataset
```

```python
independent=dataset[['id', 'age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot',
       'hemo', 'rbc', 'pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane']]

dependent=dataset[['target']]
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=42)
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators= 100, criterion="entropy")
classifier=classifier.fit(X_train,Y_train.values.ravel())
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test.values.ravel(),y_pred)
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test.values.ravel(),y_pred)
print(clf_report)
```
