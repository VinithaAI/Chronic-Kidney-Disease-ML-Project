```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
def quanQual(dataset):
        quan=[]
        qual=[]
        for columnName in dataset.columns:
            #print(columnName)
            if(dataset[columnName].dtype=='O'):
                #print("qual")
                qual.append(columnName)
            else:
                #print("quan")
                quan.append(columnName)
        return quan,qual
```

```python
quan,qual=quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df = pd.DataFrame(df,columns = quan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
cate=dataset[qual]
```

```python
cate
```

```python
cate.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(dataset[qual])
cate=imp.transform(dataset[qual])
cate = pd.DataFrame(cate,columns = qual)
```

```python
cate
```

```python
cate.isnull().sum()
```

```python
two=[df,cate]
```

```python
preprocessed=pd.concat(two,axis=1)
```

```python
preprocessed
```

```python
preprocessed.isnull().sum()
```

```python
preprocessed.to_csv("Preprocess_kidney_disease_data.csv",index=False)
```
