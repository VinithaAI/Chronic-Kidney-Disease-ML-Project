```python
independent=dataset[['id', 'age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot',
       'hemo', 'rbc', 'pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane']]
independent
```

```python
dependent=dataset[['classification']]
dependent
```

```python
dataset['classification'].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(independent, dependent)
```

```python
X_ros
```

```python
y_ros
```

```python
y_ros.value_counts()
```

```python
X_ros.shape
```

```python
# Convert the undersampled arrays to a Pandas DataFrame
Oversampled_data = pd.DataFrame(data=X_ros, columns=independent.columns)

# Add the 'dependent' variable as a new column to the DataFrame
Oversampled_data['target'] = y_ros

# Replace 'undersampled_data.csv' with the desired filename
output_file = 'Oversampled_data.csv'

# Save the DataFrame to a new CSV file
Oversampled_data.to_csv(output_file, index=False)

print(f"Oversampled data has been saved to {output_file}.")
```
